```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = [11, 10, 12, 14, 12, 15, 14, 13, 15, 102, 12, 14, 17, 19, 107, 10, 13, 12, 14, 12, 108, 12, 11, 14, 13, 15, 10, 15, 12, 10, 14, 13, 15, 10]
print(dataset)
```

```python
sns.boxplot(dataset)
```

```python
sns.histplot(dataset, kde=True)  # Adjust the number of bins as needed
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Dataset')
plt.show()
```

```python
mean = np.mean(dataset)
print(mean)
std_dev = np.std(dataset)
print(std_dev)
z_score=[(x-mean)/std_dev for x in dataset]

# Define the threshold for considering outliers (e.g., 3 standard deviations away from the mean)
threshold = 1

# Identify outliers based on z-scores
outliers = [dataset[i] for i, z_score in enumerate(z_score) if np.abs(z_score) > threshold]


print("Outliers in the dataset (beyond 3 standard deviations from the mean):")
print(outliers)
```
