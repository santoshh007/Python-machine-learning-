```python
# Step 1: Read 'imports-85.data' file using file reader
data = []
with open('/home/santosh/Desktop/Python-machine-learning-/Week 05/Day 04/datas/imports-85.data', 'r') as file:
    for line in file:
        data.append(line.strip().split(','))

# Print the first few lines of data to verify
print(data[:5])
```
