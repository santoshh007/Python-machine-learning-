```python
#y=mx+c
import pandas as pd
import matplotlib.pyplot as plt

# Define the equation parameters
m = 2  # Slope
c = 3  # Intercept

# Generate some x values
x_values = range(-10, 11)

# Calculate corresponding y values using the equation y = mx + c
y_values = [m * x + c for x in x_values]

# Create a Pandas DataFrame
data = pd.DataFrame({'x': x_values, 'y': y_values})

# Plot the data
plt.plot(data['x'], data['y'], label='y = {}x + {}'.format(m, c))
plt.xlabel('x')
plt.ylabel('y')
plt.title('Simple Linear Equation')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from the CSV file
data = pd.read_csv('./Salary_Data - Salary_Data.csv')

# Define the equation parameters
m = 2  # Slope
c = 3  # Intercept

# Calculate corresponding y values using the equation y = mx + c
data['y_predicted'] = m * data['position'] + c  # Assuming 'position' is the correct column name

# Plot the data
plt.scatter(data['position'], data['value'], label='Actual Data')  # Adjust column names accordingly
plt.plot(data['position'], data['y_predicted'], color='red', label='Predicted Line: y = {}x + {}'.format(m, c))
plt.xlabel('position')  # Adjust x-axis label
plt.ylabel('value')  # Adjust y-axis label
plt.title('Linear Regression Prediction')
plt.legend()
plt.grid(True)
plt.show()
```
